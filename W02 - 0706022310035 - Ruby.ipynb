<h1>3. Handle missing values

```python
# Assuming 'id' is the name of the column you want to drop.
# If not, replace 'id' with the actual column name
df = df.drop(columns=['url' ,'title', 'address', 'lat','long','ads_id', 'building_age'])
df.head()
```

Reason of the columns being dropped:
These columns contain unique or nearly unique text for every single rows, esp. for address, lat, long and building_age column other columns already provides a more useful  and compact information (city, year_built).

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)

null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
# An error was found in the values of 'year_built' where the year listed is 2052 [FUTURE?!]
(df['year_built'] == 2052).sum()
```

```python
# After finding that the error is only a column, its decided that the best approach is to drop the column
df = df[df['year_built'] != 2052]
```

```python
# Another duplicate data was found betweeen property_condition and furnishing
# Create a crosstab to see how the categories align
consistency_check = pd.crosstab(df['property_condition'], df['furnishing'])
print(consistency_check)
```

```python
# Based on the table above we can see that the 2 of them often contradict with each other
# The solution is to seperate them truly where all furnishing related that were in property_condition will be added 
# to the furnishing_status (a new established column)
df['furnishing_status'] = df['furnishing']

# Checking all the furnishing related rows in property_condition column
check = df['property_condition'].isin(['unfurnished', 'semi furnished'])
# Checking which rows need to be filled (if its null in furnishing status and have value related to furnishing)
rows_to_fill = df['furnishing_status'].isnull() & check
# Fill the rows
df.loc[rows_to_fill, 'furnishing_status'] = df.loc[rows_to_fill, 'property_condition']
# Replace the furnishing related stuff in property_condition to nan
df['property_condition'] = df['property_condition'].replace(['unfurnished', 'semi furnished'], np.nan)
# Delete the old column
df['furnishing'] = df['furnishing_status']
df = df.drop(columns=['furnishing_status'])
```

```python
# Fill NaN values with a suitable default (e.g., 0 or the mean) or drop them
# Using median to fill the null values [skewed]
numerical_cols = ['year_built']
for col in numerical_cols:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Using mode to fill the null values [Assuming the missing value is most likely to be the same as the most common category]
categorical_cols = ['bedrooms', 'bathrooms', 'floors', 'property_condition']
for col in categorical_cols:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)

# Fill with the safest option
df['furnishing'].fillna('unfurnished', inplace=True)
df['certificate'].fillna('nan', inplace=True)

# Drop the rows with null values in land_size_m2 and building_size_m2 (percentage in order 0.03%, 0.06%, 0.06%)
# because these are the main attributes people look at when browsing home listings. If theyâ€™re missing, the rest of the data becomes irrelevant.
df = df.dropna(subset=['land_size_m2', 'building_size_m2'])

# For Property_type all the value inside is the same (rumah) only one missing values which most likely will also be rumah. 
# And for the building orientation almost half of the data is null which can be restored but will be more inaccurate>
df = df.drop(columns=['property_type' ,'building_orientation'])
df.head()
```

```python
cols = ['bedrooms', 'bathrooms', 'carports', 'maid_bedrooms', 'maid_bathrooms', 'floors', 'year_built', 'garages']
df[cols] = df[cols].astype('int64')

# After reviewing all the unique values, the columns converted from float to int are those that do not require decimal values, 
# such as the number of floors or the number of rooms. These values are inherently whole numbers, so storing them as integers is more appropriate. 
# On the other hand, columns like price, building area and land area remain as float, since these values may reasonably contain decimals and can contain larger number (e.g., 120.5 mÂ²).
```

<h1> 4. Detect & Handle Outliers </h1>
For outliers we will check on 3 columns price_in_rp, land_seze and building size beacuse the 3 of them are the ones with the biggest range between the min and max

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of columns to check for outliers
columns_to_check = ['price_in_rp', 'land_size_m2', 'building_size_m2']

# --- 1. Visualize with Boxplots ---
print("--- Visualizing Outliers with Boxplots ---")
plt.figure(figsize=(18, 5))
for i, column in enumerate(columns_to_check):
    plt.subplot(1, 3, i + 1) # Create a subplot for each column
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# --- 2. Calculate Outliers Statistically ---
print("The Outliers (1.5 * IQR Rule)")
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count the number of outliers
    outlier_count = df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0]
    
    print(f"\nColumn: {column}")
    print(f"  - Q1: {Q1:,.2f}")
    print(f"  - Q3: {Q3:,.2f}")
    print(f"  - IQR: {IQR:,.2f}")
    print(f"  - Upper Bound for Outliers: {upper_bound:,.2f}")
    print(f"  - Number of Outliers Found: {outlier_count}")
```

For this analysis, price_in_rp, land_size_m2, and building_size_m2 have a significant number of outliers detected in the upper range of all three columns. The decision made is to keep these outliers. They are not considered data errors but a legitimate representations of the luxury segment of the real estate market. These data points correspond to large, expensive properties (i.e., estates, villas), and removing them would result in a loss of information and a biased understanding of the overall property market in the area.

<h1>5. Feature Engineering (Binning)

```python
# We'll create price categories which are 'Affordable', 'Mid-Range', 'Expensive', 'Luxury'
price_bins = [0, 1e9, 3e9, 7e9, float('inf')] # Bins in Rupiah: 0-1B, 1-3B, 3-7B, 7B+ 
price_labels = ['Affordable', 'Mid-Range', 'Expensive', 'Luxury']
df['price_category'] = pd.cut(df['price_in_rp'], bins=price_bins, labels=price_labels, right=False) 
# Right=False, means the starting value of a range is included, but the ending value is not

print("\nValue counts for the new 'price_category' feature:")
print(df['price_category'].value_counts())
```

```python
# Define the bins and labels for the year built
age_bins = [0, 1990, 2010, 2020, float('inf')]
age_labels = ['Pre-1990', '1990-2010', '2011-2020', 'Brand New']

# Create the new categorical column
df['year_built_category'] = pd.cut(df['year_built'], bins=age_bins, labels=age_labels, right=False)

print(df['year_built_category'].value_counts())
```

The 2 bins created are to help us more into categorizing the buildings based on their price and age.

<h1>Exploratory Data Analys - EDA

```python
# MOST EXPENSIVE CITY
# Calculate the median price for each city 
median_price_by_city = df.groupby('city')['price_in_rp'].median().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
sns.barplot(x=median_price_by_city.index, y=median_price_by_city.values)
plt.title('House Prices by City')
plt.xlabel('City')
plt.ylabel('Median Price in Rupiah (in Billions)')
plt.xticks(rotation=45)
plt.show()
```

This bar chart shows that Jakarta Utara and Jakarta Selatan are definitively the most expensive, premium markets, with median prices significantly higher than all other areas. There is a distinct price drop for the other Jakarta districts, followed by the surrounding cities like Tangerang, Bogor, Depok, and Bekasi, which represent the most affordable markets.

```python
# RELATION BETWEEN LAND AND BUILDING SIZES
plt.figure(figsize=(8, 6))
df['land_size_m2']
sns.scatterplot(x='land_size_m2', y='building_size_m2', data=df)
plt.title('Scatter Plot of Land sizes vs Building sizes')
plt.xlabel('land_size_m2')
plt.ylabel('building_size_m2')
plt.show()
```

The scatterplot shows a strong positive correlation between land_size_m2 and building_size_m2. As the size of the land increases, the size of the building constructed on it also tends to increase.The data is heavily clustered in the lower-left, showing that the majority of properties have both land and building sizes under 1,000 mÂ². A secondary insight is that most data points fall below a 1:1 diagonal line (the straight line), which means that homes typically have a building footprint smaller than their total land area, leaving room for a yard or garden.

```python
# RELATION BETWEEN PRICES AND BUILDING AGE CATEGORY
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='year_built_category', y='price_in_rp', order=['Pre-1990', '1990-2010', '2011-2020', 'Brand New'])
plt.title('House Price Distribution by Building Age')
plt.xlabel('Building Era')
plt.ylabel('Price in Rupiah (in Billions)')
plt.show()
```

The boxplot reveals that the median house price increases as properties get newer, with 'Brand New' homes commanding the highest median price. This suggests a strong market preference for newer constructed building. Other insight also comes from the outliers, while all building age category have high-priced outliers, the 'Brand New' and '2011-2020' categories have the most numerous and extreme high-value properties. This indicates that while a typical new home is moderately more expensive than an older one, the ultra-luxury market is almost entirely dominated by recently built ones.

```python
# IPMACT OF BUILDING AGE AND CITY ON PRICES
# Create a pivot table to summarize the median price for each combination of city and building age.
pivot_table = df.pivot_table(
    values='price_in_rp',
    index='city',
    columns='year_built_category',
    aggfunc='median'
)

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table / 1e9, annot=True, fmt='.2f') #divide by 1e9 to display the price in billions for easier reading
plt.title('Median House Price (in Billions IDR) by City and Building Age')
plt.xlabel('Building Era')
plt.ylabel('City')
plt.show()
```

The heatmap provides a detailed view of how property values are shaped by the effect of location and age. The primary insight is that the relationship between age and price is not uniform; it changes drastically depending on the city.

Contrary to a simple assumption that "newer is always more expensive," the highest median prices are found in specific niches. The most expensive properties are 'Pre-1990' homes in Bogor (25.95B IDR) and properties from the '1990-2010' era in Jakarta Pusat (24.00B IDR). This suggests a strong market for large, historic estates in Bogor and established luxury homes in Central Jakarta. In contrast, other areas like Jakarta Selatan follow a more predictable pattern where newer homes tend to be more expensive. This visualization effectively demonstrates that the definition of a "premium" property is not universal; in certain prime locations, age and prestige command a higher value than newness.

```python
# ASPECTS THAT AFFECT THE BUILDING PRICES
numerical_cols = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(14, 10))
sns.heatmap(numerical_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

The correlation matrix (heatmap) provides a quantitative summary of the linear relationships between the features of the properties, offering clear insights into the primary drivers of price. The analysis confirms that a property's physical size is the most significant factor. building_size_m2 exhibits the strongest positive correlation with price_in_rp (0.49), followed by land_size_m2 (0.37). Interestingly, these size metrics are notably stronger predictors than the room counts, with bedrooms and bathrooms showing weaker correlations. This suggests that the market values a property's overall spaciousness more than just its number of rooms.

Furthermore, the heatmap reveals the limitations of relying on a single statistical view. The year_built feature shows a very weak overall correlation with price (-0.11), which, in isolation, would suggest age is not an important factor. However, this is a misleading average that masks the complex, location-dependent trends discovered in the previous heatmap, where age was shown to have a strong but variable impact. This finding underscores the importance of using multiple visualization methods, as the correlation matrix alone cannot capture nuanced or non-linear relationships. Finally, the matrix also shows strong correlations among the predictor variables themselves (e.g., bedrooms and bathrooms), indicating that many of these features measure a single underlying concept of "property capacity."
